```python
# Import mysql connector as msql then connect mysql.
import mysql.connector as msql
from mysql.connector import Error
try:
    conn = msql.connect(host='localhost', database='bankdb', user='root', password='Redh@t#123')
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("select database();") # select the database which created all ready.
        record = cursor.fetchone()
        print("You're connected to database: ", record) # print if it's connected.
        cursor.execute('DROP TABLE IF EXISTS customer_info;') # create the table as customer_info
        print('Creating table....') # If it's done then print this
# In the below line we passed the create table's columns which we want to create
        cursor.execute("CREATE TABLE customer_info(Age int(10),Job varchar(25),Marital_Status varchar(25),Educational_Qualification varchar(25),Credit_Default varchar(10),Account_Balance int(200),Housing_Loan varchar(10),Personal_Loan varchar(10),Contact_Number varchar(20),Last_Contact_Day int(20),Last_Contact_Month varchar(20),Call_Duration int(20),Current_Campaign int(10),Term_Deposit varchar(10),Age_Group varchar(20))")
        print("Table is created....")
        # we applied a for loop through the data frame
        for i,row in dc.iterrows():
            #here %S means string values. columns and strings values should be same 
            sql = "INSERT INTO bankdb.customer_info VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            print("Record inserted")
            conn.commit() # the connection is not auto committed by default, so we must commit to save our changes
except Error as e:
            print("Error while connecting to MySQL", e)
```
